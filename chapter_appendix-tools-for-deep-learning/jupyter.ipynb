## Advanced Options

Beyond local editing two things are quite important: editing the notebooks in the markdown format and running Jupyter remotely.
The latter matters when we want to run the code on a faster server.
The former matters since Jupyter's native ipynb format stores a lot of auxiliary data that is
irrelevant to the content,
mostly related to how and where the code is run.
This is confusing for Git, making
reviewing contributions very difficult.
Fortunately there is an alternative---native editing in the markdown format.

### Markdown Files in Jupyter

If you wish to contribute to the content of this book, you need to modify the
source file (md file, not ipynb file) on GitHub.
Using the notedown plugin we
can modify notebooks in the md format directly in Jupyter.


First, install the notedown plugin, run the Jupyter Notebook, and load the plugin:

```
pip install d2l-notedown  # You may need to uninstall the original notedown.
jupyter notebook --NotebookApp.contents_manager_class='notedown.NotedownContentsManager'
```

You may also turn on the notedown plugin by default whenever you run the Jupyter Notebook.
First, generate a Jupyter Notebook configuration file (if it has already been generated, you can skip this step).

```
jupyter notebook --generate-config
```

Then, add the following line to the end of the Jupyter Notebook configuration file (for Linux or macOS, usually in the path `~/.jupyter/jupyter_notebook_config.py`):

```
c.NotebookApp.contents_manager_class = 'notedown.NotedownContentsManager'
```

After that, you only need to run the `jupyter notebook` command to turn on the notedown plugin by default.

### Running Jupyter Notebooks on a Remote Server

Sometimes, you may want to run Jupyter notebooks on a remote server and access it through a browser on your local computer. If Linux or macOS is installed on your local machine (Windows can also support this function through third-party software such as PuTTY), you can use port forwarding:

```
ssh myserver -L 8888:localhost:8888
```

The above string `myserver` is the address of the remote server.
Then we can use http://localhost:8888 to access the remote server `myserver` that runs Jupyter notebooks. We will detail on how to run Jupyter notebooks on AWS instances
later in this appendix.

### Timing

We can use the `ExecuteTime` plugin to time the execution of each code cell in Jupyter notebooks.
Use the following commands to install the plugin:

```
pip install jupyter_contrib_nbextensions
jupyter contrib nbextension install --user
jupyter nbextension enable execute_time/ExecuteTime
```

## **Utility Helpers**



1.   **check_gpu()**: Verifies GPU, CUDA, and VRAM at any point
2.   **print_trainable_parameters(model)**: Shows how many parameters are trainable (LoRA adapters vs full model)
3.   **save_adapter(model, path)**: Saves only the LoRA adapter weights

### Explanation of Each:
**transformers.AutoTokenizer**:
Converts raw text into token IDs the model understands.

**transformers.AutoModelForCausalLM**:
Loads a pre-trained causal language model (e.g., GPT-2) with a generic API.

**peft.LoraConfig & peft.get_peft_model**:
Configure and attach LoRA adapters to your base model.

**bitsandbytes.BitsAndBytesConfig**:
Sets up 4-bit quantization parameters before loading the model.

**transformers.Trainer & transformers.TrainingArguments**:
High-level API to train your model with minimal boilerplate.

**datasets.load_dataset**:
Loads datasets from Hugging Face Hub or local files, handles batching, shuffling, etc.

**torch & torch.cuda**:
Core PyTorch library for tensor operations and GPU management.

**psutil**:
Retrieves system RAM statistics.

**check_gpu():**
Prints current GPU name, VRAM, and CUDA version so you know exactly what resources you have.

**print_trainable_parameters(model)**:
Calculates the percentage of parameters that are trainable—critical to verify LoRA adapter setup.

**save_adapter(model, path)**:
Saves only the adapter weights, not the full model, so your checkpoint is small.



````python
# Imports & Utility Helpers

# 1. Core Transformers & PEFT Imports
from transformers import AutoTokenizer, AutoModelForCausalLM, Trainer, TrainingArguments
from peft import LoraConfig, get_peft_model

# 2. Dataset Loading
from datasets import load_dataset

# 3. PyTorch & System Monitoring
import torch
import psutil

# 4. Utility Functions
def check_gpu():
    """Print GPU name, total VRAM, and CUDA version."""
    if torch.cuda.is_available():
        device = torch.cuda.current_device()
        name = torch.cuda.get_device_name(device)
        total = torch.cuda.get_device_properties(device).total_memory / 1e9
        print(f"GPU: {name}, VRAM: {total:.1f} GB, CUDA: {torch.version.cuda}")
    else:
        print("No GPU available.")

def print_trainable_parameters(model):
    """Show percentage of trainable parameters in the model."""
    total_params = sum(p.numel() for p in model.parameters())
    trainable_params = sum(p.numel() for p in model.parameters() if p.requires_grad)
    pct = 100 * trainable_params / total_params
    print(f"Trainable parameters: {trainable_params:,} ({pct:.2f}% of {total_params:,})")

def save_adapter(model, path):
    """Save only LoRA adapter weights, not the full model."""
    model.save_pretrained(path, save_full_model=False)
    print(f"Adapter weights saved to {path}")
````

# **Step 4: Dataset Preprocessing**

Before training, it’s important to prepare your dataset so the model can understand it properly. This involves:

Loading your raw text data (with one sentence per entry)

Removing any unnecessary annotation fields

Tokenizing the text into input IDs that the model uses

Managing sequence length by truncation or padding

Creating labels identical to inputs for language modeling tasks

Preprocessing helps ensure smoother and efficient training on your chosen text dataset.

````python
from datasets import load_dataset

# 1. Load the dataset
dataset = load_dataset("mrmoor/cyber-threat-intelligence-relations-only")

# 2. Confirm columns
print(dataset["train"].column_names)
# → ['id', 'text', 'entities', 'relations', 'Comments']

# 3. Extract only the 'text' field and drop the rest by explicit names
def extract_text(example):
    return {"text": example["text"]}

text_only_dataset = dataset["train"].map(
    extract_text,
    remove_columns=["id", "entities", "relations", "Comments"]
)

# 4. Verify the result
print(text_only_dataset.column_names)  # → ['text']
print(text_only_dataset[0])            # → {'text': 'For example, we found ...'}

# 5. Save clean dataset for fine-tuning
text_only_dataset.to_csv("train_text.csv", index=False)
# or
text_only_dataset.to_json("train_text.jsonl", orient="records", lines=True)

print("✅ Conversion complete — text-only dataset ready.")
````
